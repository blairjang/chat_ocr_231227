```python
# 피처 각각에 대한 히스토그램을 출력
def plot_hist(df):
    plt.rcParams["figure.figsize"] = [20,16]
    fig = plt.figure(1)
    
    # df의 열 개수 만큼의 subplot을 출력
    for i in range(len(df.columns)):
        ax = fig.add_subplot(5, 5, i + 1)
        plt.hist(df[df.columns[i]], bins = 50)
        ax.set_title(df.columns[i])
        
    plt.tight_layout()
    plt.show()
```

```python
# 그래프의 단위가 일정하지 않음/ 데이터 분포가 불균형 

# 승
plot_hist(vis_pich)
```

## 데이터 분할 

```python
# 불필요한 선수명과 종속변수인 연봉을 지움 
x = pich.drop(["선수명", "연봉(2018)"], axis = 1)

# 종속변수 
y = pich["연봉(2018)"]
```

```python
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state= 10)
```

```python
x_train.shape, x_test.shape
```

### 스케일링 

```python
# 원핫 인코딩이 된 것들을 스케일링 해도 되느냐 아니냐
# 원 핫 인코딩 된 애들 : 'x0_KIA', 'x0_KT', 'x0_LG', 'x0_NC', 'x0_SK', 'x0_두산', 'x0_롯데', 'x0_삼성',
       #'x0_한화'
    
# 원핫 인코딩 안된애들을 스케일링 하자 
x_train.columns
```

```python
scale_col = x_train.columns[:19].tolist()
```

```python
# 원핫인코딩이 안된 애들을 스케일링함 
scale_col
```

```python
# 스케일링 진행
pich_ss = StandardScaler()
scaled_train = pich_ss.fit_transform(x_train[scale_col])
scaled_test = pich_ss.transform(x_test[scale_col])
```

```python
scaled_test
```

```python
len(scaled_train), len(scaled_test)
```

```python
scaled_train = pd.DataFrame(scaled_train, columns = scale_col)
scaled_test = pd.DataFrame(scaled_test, columns = scale_col)
```

```python
scaled_train.head()
```

```python
# index 가 다름 
x_train.head()
```

- pd.concat을 하면 index 가 같은 것들끼리 묶어줌 
- 하지만 스케일링 하면서 인덱스가 뒤죽박죽

```python
# x)train의 인덱스를 초기화 해줌 
scaled_train = pd.concat([scaled_train, x_train.iloc[:, 19:].reset_index(drop = True)],
                         axis = 1)
scaled_test = pd.concat([scaled_test, x_test.iloc[:, 19:].reset_index(drop = True)],
                         axis = 1)
```

```python
scaled_train.shape, scaled_test.shape
```

```python
len(y_train), len(y_test)
```

```python
# 여기까지하고 확인, 연봉까지 잘있으면서 원핫 인코딩도 잘 들어옴 
# 잘못 넣으면 null 값이 들어옴 
scaled_train.head()
```

## 모델 훈련 

```python
lr = LinearRegression()
```

```python
lr.fit(scaled_train, y_train)
```

```python
# 기울기 값이 37, 어떤애는 20340, 차이가 많이 남 
lr.coef_
```

```python
lr.intercept_
```

## 모델 평가

```python
lr.score(scaled_test, y_test)
```

```python
pred = lr.predict(scaled_test)
```

```python
mae = mean_absolute_error(y_test, pred)
mae 
```

## 모델 최적화

```python
# statsmodel 라이브러리로 회귀 분석
x = sm.add_constant(scaled_train)
```

```python
model = sm.OLS(y_train.reset_index(drop = True), x)
```

```python
model = model.fit()
```

```python
model.summary()
```

- prob(F-statistic) : F통계량에 대한 p-value. 일반적으로 0.05 이하면 회귀 분석이 유의미한 의미를 가짐
- P >|t| : 각 피처가 얼마나 유의미한지를 나타내는 p-value 

```python
# 계수 시각화 
# 회귀 계수를 시리즈로 변환
coefs = model.params.tolist()
coefs_se = pd.Series(coefs)

# 변수명을 리스트로 변환
x_labels = model.params.index.tolist()

ax = coefs_se.plot(kind = "bar")
ax.set_title("feature_coef_graph")
ax.set_xlabel("features")
ax.set_ylabel("coef")
ax.set_xticklabels(x_labels)
plt.show()
```

```python
# 피처들의 상관관계 시각화
corr = pich[scale_col].corr()

# seaborn 한글 폰트 지원 안함 
sns.set(font_scale = 1.5, font = "Malgun Gothic")
hm = sns.heatmap(corr.values, cbar = True, annot = True, fmt = ".2f",
                 annot_kws = {"size" : 15}, yticklabels = scale_col, xticklabels = scale_col)

# tight_layout : 겹치는 부분 없게 하기 
plt.tight_layout()
plt.show()

# 상관관계만 보고도 다중공산성의심 가는 것들을 찾음 
# y축의 FIP 와 KFIP 가 다중 공산성 의심 
# 이닝 - 승 ( 연관관계 강함 ) 다중 공산성 의심 
```

```python
# 피처마다 VIF 계수를 출력
# 다중 공산성이 엄청 나옴 

vif = pd.DataFrame()
vif["VIF_Factor"] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif["feature"] = x.columns
vif.round(1)
```

#### 변수제거 1

- 제거된 변수 : 볼넷/9

```python
# 볼넷 제거 후 다시 선형회귀 작업 

# AIC, BIC 체크 -> 조금 줄어듦 -> 변수제거가 유의미했다
new_x = x.drop("볼넷/9", axis = 1)
new_x = sm.add_constant(new_x)
model = sm.OLS(y_train.reset_index(drop = True), new_x)
model = model.fit()
model.summary()
```

```python
vif = pd.DataFrame()
vif["VIF_Factor"] = [variance_inflation_factor(new_x.values, i) for i in range(new_x.shape[1])]
vif["feature"] = new_x.columns
vif.round(1)
```

#### 변수제거 2

- 제거된 변수 : 볼넷/9, 승 

```python
new_x = new_x.drop("승", axis = 1)
new_x = sm.add_constant(new_x)
model = sm.OLS(y_train.reset_index(drop = True), new_x)
model = model.fit()
model.summary()

# Adj수정계수 조금 오름 
# AIC, BIC 좀 떨어짐 
# 승을 제거한 것이 옳은 선택 
```

```python
# 승, 볼넷 열이 없는지 확인 해야함 : 없음 
vif = pd.DataFrame()
vif["VIF_Factor"] = [variance_inflation_factor(new_x.values, i) for i in range(new_x.shape[1])]
vif["feature"] = new_x.columns
vif.round(1)
```

#### 변수제거 3

- 제거된 변수 : 볼넷/9, 승, 경기 

```python
new_x = new_x.drop("경기", axis = 1)
new_x = sm.add_constant(new_x)
model = sm.OLS(y_train.reset_index(drop = True), new_x)
model = model.fit()
model.summary()

# Adj수정계수 같음
# AIC, BIC 좀 떨어짐 
# 경기를 제거한 것이 옳은 선택 
```

```python
vif = pd.DataFrame()
vif["VIF_Factor"] = [variance_inflation_factor(new_x.values, i) for i in range(new_x.shape[1])]
vif["feature"] = new_x.columns
vif.round(1)
```
