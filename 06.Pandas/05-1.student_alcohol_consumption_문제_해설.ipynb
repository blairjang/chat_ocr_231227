# 문제3. 문자열 데이터를 입력 받고 capitalize() 메소드를 적용하는 capitalizer 함수 정의

```python
def capitalizer(text):
    return text.capitalize()
```

```python
capitalizer("test")
```

# 문제5. age가 17을 초과하는지 여부를 True, False로 표현하는 legal_drinker 열 생성

```python
df_sac["legal_drinker"] = df_sac["age"].map(lambda x: True if x > 17 else False)
```

```python
df_sac.head()
```

```python
# 함수로 짜면
def is_legal(x):
    if x > 17:
        return True
    else:
        return False
```

```python
df_sac["age"].map(is_legal)
```

# 문제6. int타입의 데이터에만 10을 곱하기

```python
df_sac.applymap(lambda x: 10 * x if type(x) == int else x)
```

```python
def times10(x):
    if type(x) == int:
        return 10 * x
    return x
```

```python
df_sac.applymap(times10)
```
