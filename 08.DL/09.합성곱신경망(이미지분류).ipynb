```python
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

# 데이터 준비

```python
# 패션 mnist 데이터셋 로드
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
```

# 합성곱 신경망 모델 구성

1. 합성곱 층으로 이미지에서 특징을 감지
2. 밀집층으로 클래스에 따른 분류 확률을 계산 

```python
model = keras.Sequential()
```

```python
# 합성곱 층 추가 
model.add(keras.layers.Conv2D(32, kernel_size = 3, activation = 'relu', padding = 'same',
                              input_shape = (28, 28, 1)))
```

- 패션 MNIST 이미지가 (28, 28) 크기에 세임 패딩을 적용, 32개의 필터 사용 
    - 출력되는 특성맵의 크기는 (28, 28, 32) 

```python
# 풀링층 추가 
# 채널 개수는 줄어들지 않고 크기만 14로 변함 
model.add(keras.layers.MaxPool2D(2))
```

- (2, 2) 크기의 커널을 이용하여 풀링을 적용해서 특성맵의 크기가 절반으로 줄어들고 특성맵의 깊이는 그대로 
    - 출력되는 특성맵의 크기는 (14, 14, 32) 

```python
# 두 번째 합성곱 - 풀링층 추가
# kernel_size = 3 필터의 크기는 3 ( 가로세로3 : 3, 3)
model.add(keras.layers.Conv2D(64, kernel_size = 3, activation = 'relu', padding = 'same'))
model.add(keras.layers.MaxPool2D(2))
```

- (14, 14) 크기에 세임 패딩, 64개 필터 사용 
    - 출력되는 특성맵의 크기 (14, 14, 64) 
- (2, 2) 크기의 커널을 이용하여 풀링
    - 출력되는 특성맵의 크기 (7, 7, 64) 

- 컨- 풀- 컨 - 풀 - 플레튼 - 덴스 - 드랍아웃 - 덴스 (모르면 기본)

```python
# 3차원 특성맵을 일렬로 펼침 (1차원으로)
# 이제 밀집층(덴스)에 연결
# 10개의 유닛을 가진 밀집층에서 확률을 계산하기 때문에 
# 10개인 이유 = 패션 mnist 10개라 (티셔츠인지 바지인지 등등)
model.add(keras.layers.Flatten())
```

```python
# 밀집 은닉층을 하나 더 추가한 후 드롭아웃을 넣어 과대적합을 방지하고 성능을 개선
model.add(keras.layers.Dense(100, activation = 'relu'))
model.add(keras.layers.Dropout(0.4))
```

```python
# 출력층
# 10개의 클래스를 분류하는 다중 분류 fashion MNIST 
model.add(keras.layers.Dense(10, activation = 'softmax'))
```

```python
# 모델 구조 확인
model.summary()
```

- 파라미터 개수
    - 첫 번째 합성곱 층 
        - 필터의 크기 : (3, 3, 1)    #위에서 kernel_size = 3, input_shape 마지막 1
        - 필터의 개수 : 32 
        - 3 * 3 * 1 * 32 + 32(y절편) = 320
    - 두 번째 합성곱 층 
        - 필터의 크기 : (3, 3, 32) 
        - 필터의 개수 : 64
        - 3 * 3 * 32 * 64 + 64(y절편) = 18496
        
    - 밀집 은닉 층
        - Flatten 층에서 (3136, ) 크기의 배열이 전달(7 * 7* 64)
        - 3136 * 유닛 수(100) + 유닛 수(100) = 313700 
        
    - 출력층
        - 밀집 은닉 층에서 (100, ) 크기의 배열이 전달
        - 100 * 유닛 수 10 + 유닛 수 10 = 1010

# 모델 훈련

- 옵티마이저 : Adam
- 콜백 함수 : ModelCheckpoint, EarlyStopping

```python
checkpoint_cb = keras.callbacks.ModelCheckpoint('./model/best-cnn-model.keras',
                                                save_best_only = True)
early_stopping_cb = keras.callbacks.EarlyStopping(patience = 4, restore_best_weights = True)
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
# 검증 데이터 분할
x_sub, x_val, y_sub, y_val = train_test_split(scaled_train, y_train, test_size = 0.2,stratify = y_train,random_state= 12)
```

```python
history = model.fit(x_sub, y_sub, epochs= 20, validation_data= (x_val, y_val),
                    callbacks = [checkpoint_cb, early_stopping_cb])
```
